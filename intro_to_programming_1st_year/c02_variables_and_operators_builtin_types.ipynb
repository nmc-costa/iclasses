## Variable type declarations

### Basic types

```python
string1=str("Hello")
print(type(string1), string1)
#or
string2="Hello" #just put the string
print(type(string2), string2)
```

```python
int1=int(10)
print(type(int1), int1)
#or
int2=int("10") #convert string to integer using int() function
print(type(int1), int1)
#or
int3=10 #just put the integer
print(type(int3), int3)
```

```python
float1=float(10)
print(type(float1), float1)
#or
float2=float("10") #convert string to integer using int() function
print(type(float2), float2)
#or
float3=10.0 #just put the float number
print(type(float3), float3)
```

```python
#0 or 1; 
boolean1=bool(1)
print(type(boolean1), boolean1)
boolean2=bool(0)
print(type(boolean2), boolean2)
#or
#True or False; 
boolean3=True #just put the string
print(type(boolean3), boolean3)
boolean4=False #just put the string
print(type(boolean4), boolean4)
```

### Advanced types
This will be further developed in future classes

```python
#Sequence type - sequence of  any type of values
list1=list([1,"a",3.0, True])
print(type(list1), list1)
#or
list2=[1,"a",3.0, True] #just use []
print(type(list2), list2)
```

```python
#Object type - dictionaries - object of any type of values
dict1=dict({"key1": 10, "key2": 10.0})
print(type(dict1), dict1)
#or
dict2={"key1": 10, "key2": 10.0}
print(type(dict2), dict2)
```
